```python
#convert string to list
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def data_type(text):
    list_1=[]
    for i in ast.literal_eval(text):
        list_1.append(i['name'])
    return list_1
```

```python
import_feature['genres']=import_feature['genres'].apply(data_type)
```

```python
import_feature.head()
```

```python
import_feature['keywords']=import_feature['keywords'].apply(data_type)
```

```python
import_feature.head()
```

```python
#cast count based on first 3 items
```

```python
def data_type_cast(text):
    list_1=[]
    count=0
    for i in ast.literal_eval(text):
        if count<3:
            list_1.append(i['name'])
        count+=1
    return list_1
```

```python
import_feature['cast']=import_feature['cast'].apply(data_type_cast)
```

```python
import_feature.head()
```

```python
import_feature.iloc[0]['crew']
```

```python
#fecth only director name
def data_fecth_director(text):
    list_1=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            list_1.append(i['name'])
            break
    return list_1
```

```python
import_feature['crew']=import_feature['crew'].apply(data_fecth_director)
```

```python
import_feature.head()
```

```python
import_feature['overview']
```

```python
#now convert to list
import_feature['overview']=import_feature['overview'].apply(lambda x: x.split())
```

```python
import_feature.iloc[0]['overview']
```

```python
#convert Sam Worthington to SamWorthington

def remove_space(word):
    list_1=[]
    for i in word:
        list_1.append(i.replace(" ",""))
    return list_1
```

```python
import_feature['genres']=import_feature['genres'].apply(remove_space)
import_feature['keywords']=import_feature['keywords'].apply(remove_space)
import_feature['cast']=import_feature['cast'].apply(remove_space)
import_feature['crew']=import_feature['crew'].apply(remove_space)
```

```python
import_feature.head(2)
```

# Concanate all the features

```python
import_feature['tags']=import_feature['overview']+import_feature['genres']+import_feature['keywords']+import_feature['cast']+import_feature['crew']
```

```python
import_feature.iloc[0]['tags']
```

# Converting list to string

```python
#converting list to string
str_df=import_feature[['movie_id','title','tags']]
```

```python
str_df.head(2)
```

```python
str_df['tags']=str_df['tags'].apply(lambda x:" ".join(x))
str_df.head(2)
```

```python
#converting to lower case
str_df['tags']=str_df['tags'].apply(lambda x: x.lower())
str_df.head(2)
```

```python
str_df.shape
```

# Convert text data to numerical data

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def root_word(text):
    list_1=[]
    for i in text.split():
        list_1.append(ps.stem(i))
    return " ".join(list_1)
```

```python
str_df['tags']=str_df['tags'].apply(root_word)
```

```python
str_df.sample(5)
```

```python
str_df.iloc[0]['tags']
```

```python
# convert text to number using countervectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(str_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

# cosign similarity

```python
#cosign similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity=cosine_similarity(vector)
```

```python
cosine_similarity
```

```python
cosine_similarity.shape
```

# How we get the movie index

```python
str_df[str_df['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
    index=str_df[str_df['title']== movie].index[0]
    distances=sorted(list(enumerate(cosine_similarity[index])),reverse=True, key=lambda x:x[1])
    for i in distances[1:6]:
        print(str_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
